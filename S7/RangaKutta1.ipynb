```python
# Function defining the ODE
def f(x, y):
    return x + y**2

# 4th-order Runge-Kutta method
def runge_kutta_4th_order(x, y, h):
    k1 = h * f(x, y)
    k2 = h * f(x + h/2, y + k1/2)
    k3 = h * f(x + h/2, y + k2/2)
    k4 = h * f(x + h, y + k3)
    
    y_next = y + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y_next

# Given initial values and conditions
x_0 = 0
y_0 = 1
h = 0.1
x_target = 0.2

# Number of steps required
num_steps = int((x_target - x_0) / h)

# Applying the Runge-Kutta method using a for loop
for _ in range(num_steps):
    y_0 = runge_kutta_4th_order(x_0, y_0, h)
    x_0 += h

# Print the result
print("Approximate value of y at x = 0.2:", y_0)
```

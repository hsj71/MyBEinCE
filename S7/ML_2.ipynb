```python
import string
import re
punct= []
for char in string.punctuation:
  punct.append(char)
```

```python
def cleaning(txt):

  text = txt.lower()
  text = re.sub('\s+',' ',text)
  text = text.replace("http://"," " ).replace("https://", " ")
  text = text.encode('ascii', 'replace').decode('ascii')
  text= ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
  text=  ''.join([word for word in text if word not in punct])
  text= re.sub(r"\b[a-zA-Z]\b", "",text)
  text= re.sub(r"\d+", "" , text)
  text= re.sub('\s+', ' ' , text)
  return text

```

```python
df['text_cleaned'] =df['text'].apply(lambda x: cleaning(x))
df= df[['text', 'text_cleaned', 'label']]
df.head()
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def get_wordnet_pos(word):
  """Map POS tag to first character lammetize() accepts """
  tag = nltk.pos_tag([word])[0][1][0].upper()
  tag_dict = { "J": wordnet.ADJ,
              "N" :wordnet.NOUN,
               "V":wordnet.VERB,
               "R":wordnet.ADV}
  return tag_dict.get(tag,wordnet.NOUN)
```

```python
def do_lemma(string):
  lemmatized = ' '.join([lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in nltk.word_tokenize(string)])
  return lemmatized
```

```python
df['text_cleaned'] =  df['text_cleaned'].apply(lambda x: do_lemma(x))
```
